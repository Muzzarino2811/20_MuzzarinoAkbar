2. Identifikasi kenapa menggunakan konversi warna BGR2RGB pada matplotlib dan tanpa konversi pada opencv

Perbedaan konversi warna antara OpenCV dan Matplotlib disebabkan oleh format default yang digunakan masing-masing: OpenCV menggunakan *BGR (Blue, Green, Red), sedangkan Matplotlib menggunakan **RGB (Red, Green, Blue)*./
Akibatnya, gambar yang dimuat dengan OpenCV akan tampak salah warna jika langsung ditampilkan dengan Matplotlib./
Untuk mengatasi ini, gambar harus dikonversi dari BGR ke RGB dengan cv.cvtColor(image, cv.COLOR_BGR2RGB) agar warna tampil dengan benar sesuai format Matplotlib.

```python
#SOAL 3

from skimage import io
import cv2 as cv
from google.colab.patches import cv2_imshow


image_ktp = io.imread("KTP.jpeg")
image_ktp = cv.resize(image_ktp, (0, 0), fx=0.5, fy=0.5)
image_ktp2 = cv.cvtColor(image_ktp, cv.COLOR_BGR2RGB)


height, width, _ = image_ktp2.shape


for y in range(int(height * 0.77), int(height * 0.82)):
    for x in range(int(width * 0.24), int(width * 0.5)):
        image_ktp2[y, x] = [0, 0, 255]


for y in range(int(height * 0.18), int(height * 0.68)):
    for x in range(int(width * 0.71), int(width * 0.98)):
        image_ktp2[y, x] = [255, 255, 255]


cv2_imshow(image_ktp2)
cv.waitKey(0)
cv.destroyAllWindows()
```
